```python
import datadotworld as dw
import pandas as pd
import numpy as np
import pprint as pp
import os
```

```python
#Load Meta Data from data.world
dataset = dw.load_dataset('sketchcity/city-of-houston-email-metadata-january-march-2017')
pp.pprint(dataset.describe())
```

```python
#Import data
from io import BytesIO
import csv
df = pd.read_csv(BytesIO(dataset.raw_data['original/coh_email_metadata_1Q17.csv']))
print(df.shape)
df.head()
```

```python
#Delete Null rows in relevant columns
df.dropna(subset=['Sender','To','Sent','Received'], how='any', inplace=True)
df = df.reset_index(drop=True)
print(df.shape)

#Dropping rows with incorrect times(some rows contain email ids in time columns)
df = df.drop(df.index[[x for x in range(df.shape[0]) if (df['Sent'][x][:5]!='2017-' or df['Received'][x][:5]!='2017-')]]).reset_index(drop=True)
print(df.shape)

print("Checking for Null value counts in the columns:")
print(df.isnull().sum())
```

```python
#Calculating counts of column vaues
#Sender_Counts - Number of times an emailID has sent an email
df['Sender_Counts'] = df.groupby(['Sender'])['Sender'].transform('count')
#Receiver_Counts - Number of times an emailID or group of email IDS have received an email
df['Receiver_Counts'] = df.groupby(['To'])['To'].transform('count')
#CC_Counts - Number of times an emailID or group of email IDS have received an email in CC
df['CC_Counts'] = df.groupby(['CC'])['CC'].transform('count')
#CC_Counts - Number of times an emailID or group of email IDS have received an email in BCC
df['BCC_Counts'] = df.groupby(['BCC'])['BCC'].transform('count')
#creating emails column containing by extracting number of emails in each row
df['Emails_Count'] = df[df.columns[0:4]].apply(lambda x: ','.join(x.astype(str)),axis=1)
import re
df['Emails_Count'] = df['Emails_Count'].apply(lambda x: len(re.findall(r'<(.*?)\>', x)))
```

```python
#Histogram of counts
import matplotlib.pyplot as plt
%matplotlib inline
df.hist(bins=100,figsize=(20,15))
plt.show()
```

```python
#Machine Generated Emails - heavy senders and least receivers
print("Machine Generated Emails:")
mge = df.loc[(df.Sender_Counts > 10000) & (df.Receiver_Counts < 10000/5)]
print(mge.shape)
mgelist = list(set(mge.sort_values(by=['Sender_Counts'])['Sender']))
mgelist
```

```python
#Delete machine generated from the original dataset to save time and memory for further analysis
df = df[df.Sender.str.contains('|'.join(mgelist)) == False]
df = df.reset_index(drop=True)
df.shape
```

```python
#Histogram of counts
import matplotlib.pyplot as plt
%matplotlib inline
df.hist(bins=100,figsize=(20,15))
plt.show()
```

```python
#Well connected/Trust Worthy Email IDs
print("Well connected/Trust Worthy Email IDs:")
mge = df.loc[(df.BCC_Counts > 250)&(df.Emails_Count > 5)&(df.Sender_Counts > 10)]
print(mge.shape)
wcplist = list(set(mge.sort_values(by=['BCC_Counts'],ascending=False)['BCC']))
wcplist
```

```python
#Communities/Groups (CC >10)
print("Creating subset for Communities/Groups of more than 130 people")
mge = df.loc[(df.Receiver_Counts > 1000)&(df.Sender_Counts>1000/10)&(df.Emails_Count >= 130)].sort_values(by=['Emails_Count'],ascending=True).reset_index(drop=True)
print(mge.shape)
```

```python
#creating emails column containing all email ids present in an email
mge['Emails'] = mge[df.columns[0:4]].apply(lambda x: ','.join(x.astype(str)),axis=1)
#Extracting email ids as list in each row
mge['Emails'] = mge['Emails'].apply(lambda x: re.findall(r'<(.*?)\>', x))
```

```python
# Using FP growth to find groups(frequent email groups)
import pyfpgrowth
transactions = list(mge['Emails'])
patterns = pyfpgrowth.find_frequent_patterns(transactions, 4)
rules = pyfpgrowth.generate_association_rules(patterns, 0.5)
print(patterns)
print(rules)
```
